## 数据预处理

```python
#空缺值处理
for dataset in full_data:
    dataset["Embarked"] = dataset["Embarked"].fillna("S")#观察数据发现，训练集的Embarked列中，众数为"S"，因此以此填充空缺值
    
    dataset["Fare"] = dataset["Fare"].fillna(train["Fare"].median())#由于票价存在极端值，因此用训练集中位数填补空缺值
    
    age_mean = dataset["Age"].mean()
    age_std_3 = 3 * dataset["Age"].std()
    age_null = dataset["Age"].isnull().sum()
    age_random_list = np.random.randint(age_mean - age_std_3,age_mean + age_std_3,size = age_null)
    dataset["Age"][np.isnan(dataset["Age"])] = age_random_list#在均值附近的三倍标准差之内产生随机数，填补空缺值
    
    dataset["Cabin"] = dataset["Cabin"].fillna("N")#由于缺失比例过大，直接将空缺值作为新的一类
test.info()
```

## 数据编码

```python
train.describe(include = "all")
```

```python
CategoricalSex_map = {"male":1,"female":0}
CategoricalEmbarked_map = {"C":0,"Q":1,"S":2}
for dataset in full_data:
    dataset["CategoricalSex"] = dataset["Sex"].map(CategoricalSex_map).astype(int)
    
    dataset["CategoricalCabin"] = 1
    dataset.loc[dataset["Cabin"] == "N",["CategoricalCabin"]] = 0
    
    dataset["CategoricalEmbarked"] = dataset["Embarked"].map(CategoricalEmbarked_map).astype(int)

train.describe(include = "all")  
```

```python
drop_feature_list = ["PassengerId","Name","Sex","Ticket","Fare","Cabin","Embarked",
                     "FamilySize"]
train = train.drop(drop_feature_list,axis = 1)
test = test.drop(drop_feature_list,axis = 1)
train.head()
```

## 建模

### 随机森林

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest = train_test_split(train.iloc[:,1:],train.iloc[:,0],random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
rf = RandomForestClassifier(oob_score = True,
                            random_state = 42)
param_grid = {"criterion":["gini","entropy"],
             "min_samples_split" : [2,4,6,8,10,12],
             "min_samples_leaf" : list(range(1,12,5)),
             "n_estimators" : list(range(10,101,20))}
gs = GridSearchCV(estimator = rf,
                  param_grid = param_grid,
                  scoring = "accuracy",
                  cv = 3)
gs.fit(Xtrain,ytrain)
```

```python
print(gs.best_score_)
print(gs.best_params_)
```

```python
rf_1 = RandomForestClassifier(criterion = "entropy",
                              min_samples_leaf = 1,
                              min_samples_split = 8,
                              n_estimators = 10,
                              oob_score = True,
                              random_state = 42)
rf_1.fit(Xtrain,ytrain)
print(rf_1.oob_score_)
```

```python
pd.concat((pd.DataFrame(rf_1.feature_importances_,columns = ["importance"]),
          pd.DataFrame(Xtrain.columns,columns = ["features"])),
          axis = 1).sort_values(by = "importance",ascending = False)[:20]
```

## 预测

```python
pred_results = rf_1.predict(test)
prediction = pd.DataFrame(pred_results,columns = ["Survived"])
test_0 = pd.read_csv("test.csv")
results = pd.concat((test_0.iloc[:,0],
                     prediction),
                    axis = 1)
results.to_csv(("result.csv"),sep=",", index = False)
```
